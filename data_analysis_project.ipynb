```python
path = '/content/drive/MyDrive/melting_tank.csv'
```

```python
df = pd.read_csv(path) # 공정데이터 분석
df
```

```python
df.columns = df.columns.str.lower() # 분석용이하게 열제목 소문자화
df.columns
```

```python
df.isnull() # 결측치 확인과 결측치를 표시하는 dataframe 생성
```

```python
df.isnull().sum() # 열별로 결측치 개수 count
```

```python
df= df.drop(columns=['num']) # num 없애
df.head(2) # 확인차 두줄 뽑아보자. why? 시간절약
```

```python
df = df.dropna(how='any', axis=0) # dr.dropna로 NA(결측치)제거
df
```

```python
df['std_dt'] = pd.to_datetime(df['std_dt']) # 문자열을 'datetime'자료형으로 변환
df.info()
```

```python
df['month'] = df['std_dt'].dt.month.astype('str') # 'datetime'형식으로 바꾼 'std_dt'를 month, day, hour로 각각 뽑아서 열로 만들어줌.
df['day'] = df['std_dt'].dt.day.astype('str')
df['hour'] = df['std_dt'].dt.hour.astype('str')
```

month_day의 경우 df.month함수를 이용하여 만들 수 있지만.
3/4,3/5.....4/1,4/2.....4/12 등으로 데이터가 생성된다면
41,4/11,4/12 순서로 정렬되므로 04/01,04/02와 같은 방식으로 데이터를 생성
df['month_day'] = (df['std_dt'].astype('str').str.split('_').str[1] + '/' + df['std_dt'].astype('str).str.split('-').str[2].str.split('').str[0])
df

```python
df = df[['std_dt', 'month_day','melt_temp', 'motorspeed', 'melt_weight', 'insp', 'tag',
       'month', 'day', 'hour']] # 새로 정의된 'month', 'day', 'hour'새로운 dataframe 생성
df.head()
```

```python
col_name = ['melt_temp','motorspeed','melt_weight','insp'] # 시간에 따른 네 가지 항목값을 보자

for i in range(len(col_name)):
  plt.figure(figsize=(12,5))
  plt.plot(df.std_dt,df[col_name[i]]) # 날짜에 따른 항목별 값
  plt.title(col_name[i])
  plt.show() # 그러나 너무 촘촘해서 보기힘들다.
```

```python
for i in range(len(col_name)): 
  plt.figure(figsize=(12,5))
  plt.plot(df[col_name[i]][0:1000]) # 위 그래프 0~1000까지 범위만 촘촘하지 않게 보자
  plt.title(col_name[i])
  plt.show() 
```

```python
df.hour.unique()
```

```python
month_day = df.month_day.unique()[0] # 첫날 즉, 03/04
hour='1'
section_df = df[(df.month_day == month_day) & (df.hour == hour)]
section_df # 03/04 한시 dataframe
```

```python
group = df.groupby(['month_day','tag'])['tag'].count() # 날짜별로 tag개수 세자
group = pd.DataFrame(group)
group = group.rename(columns = {'tag':'count'})
group = group.reset_index()
group
```

```python
plt.figure(figsize = (20,20
                      ))
sns.barplot(x='month_day', y='count', hue = 'tag',data=group) # 날짜별 불량분포
plt.show()
```

```python
col_name = ['melt_temp', 'motorspeed', 'melt_weight', 'insp']
```

```python
plt.figure(figsize=(10,10))

for i in range(len(col_name)): # 공정조건별 불량분포
  num = 1+i
  plt.subplot(2,2,num)
  sns.boxplot(x='tag',y=col_name[i],data=df)
  plt.title(col_name[i])
plt.show()
```

```python
plt.figure()
sns.scatterplot(x='melt_temp',y='motorspeed',hue='tag',data=df)
plt.show() # 먼저찎힌 점이 가려짐
```

```python
# 그룹을 따로 나누어서 overlap으로 그리면 됩니다.
df_ok = df[df['tag']=='OK']
df_ng = df[df['tag']=='NG']
# OK 인거 먼저그리고 NG 인거를 나중에 그리면 NG를 파악하기 좋다.
plt.figure()
sns.scatterplot(x='melt_temp',y='motorspeed', data=df_ok, label='temp')
sns.scatterplot(x='melt_temp',y='motorspeed', data=df_ng, label='speed')
plt.legend()
plt.show() # 낮은'temp'와 낮은'speed'에서 'NG'가 난다.
```

```python
df_sort = df.sort_values(by = ['tag'], ascending=False)

plt.figure()
sns.scatterplot(x='melt_temp', y='motorspeed',hue='tag',data=df_sort)
plt.show()
```
